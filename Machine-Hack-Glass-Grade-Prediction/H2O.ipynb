```python
from h2o.automl import H2OAutoML
```

```python
train = h2o.import_file('Train.csv')
test = h2o.import_file('Test.csv')
```

```python
train.describe()
```

```python
train['class'] = train['class'].asfactor()
train['class'].table()
```

```python
x_train, x_test, x_valid = train.split_frame(ratios=[.85,0.10],seed=11)
x_train.shape, x_test.shape, x_valid.shape
```

```python
y = 'class'
X = train.columns
X.remove(y)
```

```python
aml = H2OAutoML(nfolds=10,
    balance_classes=True,
    class_sampling_factors=None,
    max_after_balance_size=5.0,
    max_runtime_secs=None,
    max_runtime_secs_per_model=None,
    max_models=None,
    stopping_metric='AUTO',
    stopping_tolerance=None,
    stopping_rounds=None,
    seed=4,
    project_name=None,
    exclude_algos=['DeepLearning'],
    include_algos=None,
    exploitation_ratio=0,
    modeling_plan=None,
    monotone_constraints=None,
    algo_parameters=None,
    keep_cross_validation_predictions=True,
    keep_cross_validation_models=False,
    keep_cross_validation_fold_assignment=True,
    sort_metric='logloss',
    export_checkpoints_dir=None,
    verbosity='warn',
)
```

```python
aml.train(x = X, y = y, training_frame = x_train, validation_frame=x_valid)
```

```python
lb = aml.leaderboard
lb
```

```python
pred_test = aml.leader.predict(test)
```

```python
aml.leader.confusion_matrix
```

```python
h2o.export_file(pred_test, path ='sub.csv', force = True)
```
